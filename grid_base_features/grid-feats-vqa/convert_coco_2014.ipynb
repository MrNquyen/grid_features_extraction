# **FUNCTION**

```python
# Load json
def load_json(path):
    with open(path, 'r', encoding='utf-8') as file:
        json_file = json.load(file)
        return json_file
    

# Save json
def save_json(path, dic):
    with open(path, 'w', encoding='utf-8') as file:
        json.dump(dic, file, indent=3, ensure_ascii=False)
```

# **FORMAT COCO 2014**

```python
image_path = f'{images_path}\\0002762.png'

img = io.imread(image_path)
img.shape
```

```python
# Image format
item_images_template = {
    'file_name': None,
    'id': None,
    'height': None,
    'width': None,
}

# Coco2014 full format
coco_format = {
    'images': [], 
    # 'annotations': [], 
    # 'categories': [], 
    # 'attCategories': [],
}


def format_coco(annotation_file):
    coco_annotation = coco_format.copy()
    item_image = item_images_template.copy()
    item_annotations = {}
    item_categories = {}
    item_attCategories = {}

    for id in annotation_file.keys()[:10]:
        file_name = f'{id}.png'
        image_path = f'{images_path}\\{id}.png'
        img = io.imread(image_path)
        height, width = img.shape[:2]

        item_image = {
            'file_name': file_name,
            'id': id,
            'image_id': id,
            'height': height,
            'width': width,
        }

        coco_annotation['images'].append(item_image)
        # coco_annotation['annotations'].append(item_annotations)
        # coco_annotation['categories'].append(item_categories)
        # coco_annotation['attCategories'].append(item_attCategories)

    return coco_annotation
```

```python
train_coco = format_coco(train_annotation)
val_coco = format_coco(val_annotation)
test_coco = format_coco(test_annotation)
```

```python
save_path = 'F:\\UNIVERSITY\\NCKH\\NCKH_V2\\code\\feature_extraction\\grid_base_features\\grid-feats-vqa\\datasets\\coco\\annotations'
save_json(f'{save_path}\\instances_train2014.json', train_coco)
save_json(f'{save_path}\\instances_val2014.json', val_coco)
save_json(f'{save_path}\\instances_test2014.json', test_coco)
```
